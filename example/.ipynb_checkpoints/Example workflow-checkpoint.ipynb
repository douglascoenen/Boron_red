```python
# Initialise the class for the analysis
# It contains all the methods that will reduce the data step by step with diagnostic graphs along the way
# When call with no options it will use Tkinter (GUI library) to prompt you to pick your files, otherwise they can
# manually entered as strings.
red = Reduced()
```

# .logtime(check, manstart) method

Will take in the raw .exp file (from Neptune plus) and .csv log file (from Geostar).

### The two options are:
*check* (bool) => set to True (default) if you want to plot some diagnostic plots to see if the automatic detection worked. Works better with qt activate 

*manstart* (bool) => set to True if you want to manually select the start of the first analysis in case the automatic detection did not work. This will then calculate the laser - MS time offset and apply it to all analysis
                      

```python
red.logtime(True)
```

# .bgsub(mancut, savefig, endcut) method

Will go through each analysis and substract the background to the signal and calculate the raw ratios

### The three options are:
*mancut* (bool) => set to True if you want to manually cut each sample **individually** (very lengthy). Left click at start and end of where you want to cut. **Qt** backend necessary.

*savefig* (bool) => set to True if you want to save a plot of each analysis (**recommended**). Very useful to diagnose outliers. Works for both automatic and manual cutting. It will save each analysis as a file in a subfolder. 

*endcut* (int) => number of integration cycles to remove from the end of the signal. If the automatic cutting from .logtime() did not go as planned, you can shorten it.

```python
red.bg_sub()
```

# .nistcor(mancut, savefig, endcut) method

NIST612 correction. Will use the nists as bracketing standard to estimate the mass bias and correct the measured ratios. Will also estimate [B] by scaling NIST612 11B voltage to sample 11B voltage. It will also estimate B/Ca by using 11B/10.035 ratio as proxy for B/Ca using JCp, JCt or MACS-3 as bracketing standard.

### The two options are:
*save* (bool) => Save the diagnostic plots

*poly* (bool) => Instead of using bracketing NISTs, use a polynomial to fit through all the NISTs of the session **experimental**.


```python
red.nistcor()
```

# .ca_cor(outlier, save, ca9, yel, spot, flu)

The main part of the script. Will use MACS-3, JCt and JCp to create a inaccuracy calibration line for the session.

It will automatically remove outliers based on the residuals of the fit (first figure)

### Options

*outlier* (bool) => set to True to manually select outliers (**not recommended**)

*save* (bool) => set to True (default) to save the figures.

*ca9* (bool) => set to True to use the cup 9.979 instead of 10.035.

*yel* (bool) => set to True to use our yellow calcite standard for really low concentration samples. 

*spot* (int) => restrain the dataset to a specific spot size to perform the calibration.

*flu* (float) => restrain the dataset to a specific fluence.


Some attempts were made to make a Monte Carlo fitting (to use the uncertainty of each point) and yields very similar results. A multiple regression using both 9.979 and 10.035 was also attempted but nothing convicing came out of it, mostly several permil offsets.

```python
red.ca_cor()
```

# .uwc_plot(save, name)

Will plot our main secondary standards to check the accuracy of the session.

### Options

*save* (bool) => set to True (default) to save the figures.

*name* (str) => change  to change the file name under which the plots are saved, if you want to try different settings in the previous methods

```python
red.uwc_plot()
```

# .savedata(name)

Will save the main dataframe as a single csv file (*altogether.csv*). 

It will separate the samples from the standards (which were detected): *standards.csv*, *samples.csv*

It will also save each standard as a single csv file for convenience in a subfolder.

In case you want to reuse the previous **datadi** dictionary (especially if you did some painstaking manual cutting), it is saved as a pickle file and can be reloaded


### Options

*outlier* (bool) => set to True to manually select outliers (**not recommended**)

*save* (bool) => set to True (default) to save the figures.

*ca9* (bool) => set to True to use the cup 9.979 instead of 10.035.

*yel* (bool) => set to True to use our yellow calcite standard for really low concentration samples. 

*spot* (int) => restrain the dataset to a specific spot size to perform the calibration.

*flu* (float) => restrain the dataset to a specific fluence.


Some attempts were made to make a Monte Carlo fitting (to use the uncertainty of each point) and yields very similar results. A multiple regression using both 9.979 and 10.035 was also attempted but nothing convicing came out of it, mostly several permil offsets.
